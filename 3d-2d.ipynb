```python
import cv2
import numpy as np
def nothing(x):
    pass
blank = np.zeros((600,600,3), np.uint8)
n=0
cam= cv2.VideoCapture(0)
cv2.namedWindow('image')
cv2.createTrackbar('hmin','image',0,255,nothing)
cv2.createTrackbar('smin','image',0,255,nothing)
cv2.createTrackbar('vmin','image',0,255,nothing)
cv2.createTrackbar('hmax','image',0,255,nothing)
cv2.createTrackbar('smax','image',0,255,nothing)
cv2.createTrackbar('vmax','image',0,255,nothing)

while True:
    hmin = cv2.getTrackbarPos('hmin','image')
    smin = cv2.getTrackbarPos('smin','image')
    vmin = cv2.getTrackbarPos('vmin','image')
    hmax = cv2.getTrackbarPos('hmax','image')
    smax = cv2.getTrackbarPos('smax','image')
    vmax = cv2.getTrackbarPos('vmax','image')
    
    returnVal,frame=cam.read()
    xx=cv2.GaussianBlur(frame,(5,5),0)
    xx=cv2.cvtColor(xx,cv2.COLOR_RGB2HSV)
    min=np.array([hmin, smin, vmin],np.uint8)
    max=np.array([hmax,smax,vmax],np.uint8)
    obj=cv2.inRange(xx,min,max)
    cover = cv2.GaussianBlur(obj, (5,5),0)
    moments = cv2.moments(cover)
    m00=moments['m00']
    key = cv2.waitKey(15)
    if m00 != 0:
     cx = int(moments['m10']/m00)
     cy = int(moments['m01']/m00)
     if key == 32 and n==0:
      n=1   
     if n==1:
      cv2.circle(blank,(cx,cy),12,(255,255,255),-1)
     if key == 119 and n==1:
      n=0   
     if n==0:
      cv2.circle(blank,(cx,cy),12,(0,0,0),-1)
     print(cx,cy,n) 
    else:
     print(m00)
    oj=cv2.flip(obj,1)
    blan=cv2.flip(blank,1)
    frame=cv2.flip(frame,1)
    cv2.imshow('final',blan)
    cv2.imshow('finalc',oj)
    cv2.imshow('final0',frame)
    
    if key == 27: break
```
